```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pylab as pil
```

```python
sns.set()
```

```python
gr = np.loadtxt('test_GR_scalCls.dat')
mg = np.loadtxt('test_GR_mugamma_scalCls.dat')
qr = np.loadtxt('test_GR_QR_scalCls.dat')
```

```python
plt.semilogx(gr[:,0], gr[:,1], color = 'C2')
plt.semilogx(mg[:,0], mg[:,1], color = 'C0')
plt.semilogx(qr[:,0], qr[:,1], color = 'C1')
plt.xlim(2,2500)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$\ell(\ell + 1) C_{\ell}^{\rm TT} / 2 \pi$')
plt.grid(b=True, which='minor', color='white')
plt.show()
```

```python
plt.semilogx(gr[:,0], (mg[:,1]-gr[:,1])/gr[:,1] )
plt.semilogx(gr[:,0], (qr[:,1]-gr[:,1])/gr[:,1] )
#plt.plot(gr[:,0], (fr[:,1]-gr[:,1])/gr[:,1] )
plt.xlim(2,2500)
plt.ylim(-0.0005,0.0005)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$ \Delta C_{\ell}^{\rm TT} / C_{\ell}^{\rm TT, GR} $ ')
plt.grid(b=True, which='minor', color='white')
pil.savefig('delta_cl_mgcamb.pdf', bbox_inches='tight')
plt.show()
```

```python
gr_m = np.loadtxt('test_GR_matterpower.dat')
mg_m = np.loadtxt('test_GR_mugamma_matterpower.dat')
qr_m = np.loadtxt('test_GR_QR_matterpower.dat')
```

```python
plt.loglog(gr_m[:,0], gr_m[:,1], color = 'C2')
plt.loglog(mg_m[:,0], mg_m[:,1], color = 'C0')
plt.loglog(qr_m[:,0], qr_m[:,1], color = 'C1')
plt.xlim(0.0001,1)
plt.xlabel(r'$k \, [h/{\rm Mpc}]$')
plt.ylabel(r'$P(k)$')
plt.grid(b=True, which='minor', color='white')
plt.show()
```

```python
plt.semilogx(gr_m[:,0], (mg_m[:,1]-gr_m[:,1])/gr_m[:,1], color = 'C0' )
plt.semilogx(gr_m[:,0], (qr_m[:,1]-gr_m[:,1])/gr_m[:,1], color = 'C1' )
#plt.plot(gr[:,0], (fr[:,1]-gr[:,1])/gr[:,1] )
plt.xlim(0.0001,1)
plt.ylim(-0.0005,0.0005)
plt.xlabel(r'$k \, [h/{\rm Mpc}]$')
plt.ylabel(r'$ \Delta P(k) / P(k)_{\rm GR} $ ')
plt.grid(b=True, which='minor', color='white')
pil.savefig('delta_pk_mgcamb.pdf', bbox_inches='tight')
plt.show()
```
