```python
k1 = 0.10112973E-03
k2 = 0.10910861E-02
k3 = 0.10661010E-01
k4 = 0.10273624E+00
```

```python
dfc_1 = dfc.loc[dfc['k'] == k1]
dfm_1 = dfm.loc[dfm['k'] == k1]
dfm2_1 = dfm2.loc[dfm2['k'] == k1]

dfc_2 = dfc.loc[dfc['k'] == k2]
dfm_2 = dfm.loc[dfm['k'] == k2]
dfm2_2 = dfm2.loc[dfm2['k'] == k2]

dfc_3 = dfc.loc[dfc['k'] == k3]
dfm_3 = dfm.loc[dfm['k'] == k3]
dfm2_3 = dfm2.loc[dfm2['k'] == k3]

dfc_4 = dfc.loc[dfc['k'] == k4]
dfm_4 = dfm.loc[dfm['k'] == k4]
dfm2_4 = dfm2.loc[dfm2['k'] == k4]
```

```python
# Defining the plotting functions
def plot(quantity):
    plt.semilogx(dfc_3['a'].values,  dfc_3[quantity].values,  color = 'black', linestyle = '--', label= 'CAMB')
    plt.semilogx(dfm_3['a'].values,  dfm_3[quantity].values,  color = 'C0', label= r'MGCAMB, $a_{\rm GR} = 10^{-2}$')
    plt.semilogx(dfm2_3['a'].values, dfm2_3[quantity].values, color = 'C1', label= r'MGCAMB, $a_{\rm GR} = 10^{-3}$')

    
    plt.xlabel(r'$a$')
    plt.legend(loc='lower right')
    plt.ylabel(quantity)
    plt.xlim(5e-4,1.1)
    plt.title(quantity)
    plt.show()
```

```python
def plot_diff(quantity):
    #plt.semilogx(dfc_2['a'].values,  dfc_2[quantity].values,  color = 'black', linestyle = '--', label= 'CAMB')
    plt.semilogx(dfm_2['a'].values,  (dfm_2[quantity].values-dfc_2[quantity].values)/dfc_2[quantity].values,  color = 'C0', label= r'MGCAMB, $a_{\rm GR} = 10^{-2}$')
    plt.semilogx(dfm2_2['a'].values, (dfm2_2[quantity].values-dfc_2[quantity].values)/dfc_2[quantity].values, color = 'C1', label= r'MGCAMB, $a_{\rm GR} = 10^{-3}$')

    
    plt.xlabel(r'$a$')
    plt.legend(loc='lower left')
    plt.ylabel(quantity)
    plt.xlim(5e-4,1.1)
    plt.ylim(-0.01,0.01)
    plt.title(quantity)
    plt.show()
```

```python
plot('pidot_sum')
```

```python
plot_diff('pidot_sum')
```

```python
plot('dgpi')
```

```python
plot_diff('dgpi')
```

## CMB power spectrum differences

### $N_{\nu} = 1$, $m_{\nu} = 0.06$ eV

```python
errors_cmb = []
dev_errors_cmb = []
mnu = [0.06, 0.12, 0.18, 0.6]
```

```python
gr = np.loadtxt('./LCDM_std_mnu_0p06_scalCls.dat')
mg = np.loadtxt('./LCDM_mg_mnu_0p06_grt_0p01_scalCls.dat')
ms = np.loadtxt('./LCDM_ms_mnu_0p06_grt_0p01_scalCls.dat')
qr = np.loadtxt('./LCDM_qr_mnu_0p06_grt_0p01_scalCls.dat')
```

```python
errors_cmb.append(np.sqrt(((mg[:,1]-gr[:,1])/gr[:,1]*100)**2).mean())
dev_errors_cmb.append(np.sqrt(np.sqrt(((mg[:,1]-gr[:,1])/gr[:,1]*100)**2).var())
```

```python
plt.semilogx(mg[:,0], (mg[:,1]-gr[:,1])/gr[:,1]*100, linestyle = '--', linewidth = 2, label=r'$\mu, \gamma$')
plt.semilogx(ms[:,0], (ms[:,1]-gr[:,1])/gr[:,1]*100, linestyle = '-.', linewidth = 1.5,label=r'$\mu, \Sigma$')
plt.semilogx(qr[:,0], (qr[:,1]-gr[:,1])/gr[:,1]*100, linestyle = '-', linewidth = 0.75,label=r'$Q,R$')
plt.legend(loc='upper right')
plt.xlabel(r'$\ell$', fontsize=13)
plt.ylabel(r'$\Delta C_{\ell} / C_{\ell}^{\rm CAMB}$ [%]', fontsize=13)
plt.xlim(2,2200)
plt.ylim(-0.005,0.005)
plt.title(r'$m_{\nu} = 0.06$ eV')
pil.savefig('./mgcamb_plots/LCDM_mnu_0p06.pdf', bbox_inches = 'tight')
plt.show()
```
